<p>Import drive and use OpenCV to load the image</p>

```python
img = cv2.imread('/content/drive/MyDrive/Mod4/cv_logo.png')
# the line below is necessary because OpenCV uses BGR format
# and matplotlib reads images using RGB format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

<p>The kernel is defined by 1/(K_width * K_height)</p>
<li>So, for a 3x3 kernel matrix, we would specify the following:</li>

```python
kernel = np.ones((3, 3), np.float32) / 9
```

<h4>Destination Image</h4>
<li>first argument is the source image, seocnd is the desired depth, 3rd is the kernel</li>

```python
dst = cv2.filter2D(img, -1, kernel)
titles = ['image', '2D Convolution']
images = [img, dst]
```

<h4>Plot Images</h4>

```python
for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```

<li>We can see the original image and the destination image</li>
<li>If done correctly, the image should be slightly blurred, and some of the imperfections in the original image around the sharp edges should be smoothed</li>

<h4>Low and High Pass Filtering</h4>
<li>LPF - helps in removing noise, blurring images</li>
<li>HPF - helps in finding edges in the images</li>

```python
blur = cv2.blur(img, (3, 3))
titles = ['images', '2D Convolution', 'blur']
images = [img, dst, blur]

for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks()
    plt.yticks()
```

<h4>Gaussian Filtering with 3x3 Kernel</h4>
<li>Gaussian filtering, simply put, uses a different kernel in both x and y direction</li>
<li>The kernel matrix has the highest value in the middle and decreases as we traverse outward</li>

```python
img = cv2.imread('/content/drive/MyDrive/Mod4/Halftone_Gaussian_Blur.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kernel = np.ones((3, 3), np.float32) / 9
dst = cv2.filter2D(img, -1, kernel)

blur = cv2.blur(img, (3, 3))
g_blur = cv2.GaussianBlur(img, (3, 3), 0)
titles = ['images', '2D Convolution', 'blur', 'Gaussian Blur']
images = [img, dst, blur, g_blur]

for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks()
    plt.yticks()

plt.show()
```

<h4>Comparing Filters</h4>

*   Mean filtering replaces each pixel's value with the median of its neighboring pixels. 
*   This typically works well for salt and pepper images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/MyDrive/Mod4/Mod4CT1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel_3 = np.ones((3, 3), np.float32) / 9
median_blur_k3 = cv2.medianBlur(img, 3)
mean_k3 = cv2.blur(img, (3, 3))
gaussian_blur_k3_sigma0 = cv2.GaussianBlur(img, (3, 3), 0)
gaussian_blur_k3_sigma3 = cv2.GaussianBlur(img, (3, 3), 3)

kernel_5 = np.ones((5, 5), np.float32) / 25
median_blur_k5 = cv2.medianBlur(img, 5)
mean_k5 = cv2.blur(img, (5, 5))
gaussian_blur_k5_sigma0 = cv2.GaussianBlur(img, (5, 5), 0)
gaussian_blur_k5_sigma3 = cv2.GaussianBlur(img, (5, 5), 3)

kernel_7 = np.ones((7, 7), np.float32) / 49
median_blur_k7 = cv2.medianBlur(img, 7)
mean_k7 = cv2.blur(img, (7, 7))
gaussian_blur_k7_sigma0 = cv2.GaussianBlur(img, (7, 7), 0)
gaussian_blur_k7_sigma3 = cv2.GaussianBlur(img, (7, 7), 3)

titles = ['Median', 'Mean', 'Gauss s0', 'Gauss s3']
kernel_titles = ['K3', 'K5', 'K7']
images = [
    median_blur_k3, mean_k3, gaussian_blur_k3_sigma0, gaussian_blur_k3_sigma3,
    median_blur_k5, mean_k5, gaussian_blur_k5_sigma0, gaussian_blur_k5_sigma3,
    median_blur_k7, mean_k7, gaussian_blur_k7_sigma0, gaussian_blur_k7_sigma3
]

for i in range(12):
    plt.subplot(3, 4, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i % 4] + " " + kernel_titles[i // 4])
    plt.xticks([])
    plt.yticks([])

plt.show()
```
